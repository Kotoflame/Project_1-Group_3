```python
# Import the Data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the Data
df_eggs = pd.read_csv('C:/Users/nelso/Documents/Project_1-Group_3/Resources/API_Data/Cleaned_Eggs_2013.csv')
```

```python
# Explore the Data
df_eggs.head()
```

```python
# Create a histogram of the data
sns.histplot(data=df_eggs, x='3-Month % Change', kde=True)

# Calculate and print the standard deviation of the data
std_dev = np.std(df_eggs['3-Month % Change'])
print(f"The standard deviation of the data is: {std_dev:.2f}")
```

```python
# Calculate and print the mean of the data
mean = np.mean(df_eggs['3-Month % Change'])
print(f"The mean of the data is: {mean:.2f}")

# Calculate and print the standard deviation of the data
std_dev = np.std(df_eggs['3-Month % Change'])
print(f"The standard deviation of the data is: {std_dev:.2f}")
```

```python
# Create a box plot of the data
sns.boxplot(data=df_eggs, x='3-Month % Change')

# Add a title to the plot
plt.title('Box Plot of 3-Month % Change')

# Show the plot
plt.show()
```
