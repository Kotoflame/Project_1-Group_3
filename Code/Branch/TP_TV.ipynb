```python
#import functions

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import csv TV 

tv_df = pd.read_csv("../Resources/API_Data/Cleaned_TV_2013.csv")
tv_df
```

```python
#Get average value per year
yearly_avg = tv_df.groupby('Year')['Value'].mean()
```

```python
# Dataframe for TV Value

# Convert 'Date' column to datetime format
tv_df['Date'] = pd.to_datetime(tv_df['Date'])


# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the '3-Month % Change' column as a line plot
ax.plot(tv_df['Norm TV Values'], linewidth=1.5, color='blue')


# Add labels and a title
ax.set_xlabel('Date')
ax.set_ylabel('3-Month % Change')
ax.set_title('TV Value Change over Time')

# Display the graph
plt.show()
```

```python
# Dataframe for TV 3 Month % change

# Convert 'Date' column to datetime format
tv_df['Date'] = pd.to_datetime(tv_df['Date'])

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the '3-Month % Change' column as a line plot
ax.plot(tv_df['3-Month % Change'], linewidth=1.5, color='blue')


# Add labels and a title
ax.set_xlabel('Date')
ax.set_ylabel('3-Month % Change')
ax.set_title('TV 3-Month Change over Time')

# Display the graph
plt.show()
```

```python
# import csv TP
tp_df = pd.read_csv("../Resources/API_Data/Cleaned_TP_2013.csv")
tp_df
```

```python
#Get average value per year
yearly_avg = tp_df.groupby('Year')['Value'].mean()
```

```python
# DataFrame for Toilet Paper Value

# Convert 'Date' column to datetime format
tp_df['Date'] = pd.to_datetime(tp_df['Date'])

# Set 'Date' column as the index
tp_df = tp_df.set_index('Date')

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the '3-Month % Change' column as a line plot
ax.plot(tp_df['Value'], linewidth=1.5, color='blue')

# Set the y-axis limits and ticks
#ax.set_ylim(-6, 6)
#ax.set_yticks(range(-6, 7, 2))

# Add labels and a title
ax.set_xlabel('Date')
ax.set_ylabel('3-Month % Change')
ax.set_title('Toliet Paper Value Change over Time')

# Display the graph
plt.show()
```

```python
# DataFrame for Toilet Paper 3-Month % change

# Convert 'Date' column to datetime format
tp_df['Date'] = pd.to_datetime(tp_df['Date'])

# Set 'Date' column as the index
tp_df = tp_df.set_index('Date')

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the '3-Month % Change' column as a line plot
ax.plot(tp_df['Value'], linewidth=1.5, color='blue')

# Set the y-axis limits and ticks
#ax.set_ylim(-6, 6)
#ax.set_yticks(range(-6, 7, 2))

# Add labels and a title
ax.set_xlabel('Date')
ax.set_ylabel('3-Month % Change')
ax.set_title('Toilet Paper 3-Month % Change over Time')

# Display the graph
plt.show()
```

```python
# plot both TV and TP Value 

fig, ax = plt.subplots(figsize=(10, 5))

# plot TV data as red line
ax.plot(tv_df['Year'], tv_df['Value'], color='red', label='TV')

# plot TP data as blue line
ax.plot(tp_df['Year'], tp_df['Value'], color='blue', label='TP')

# add title and axis labels
ax.set_title('TV and Toilet Paper Value Change by Year')
ax.set_xlabel('Year')
ax.set_ylabel('3-Month % Change')

# add legend
ax.legend()

# display the plot
plt.show()






```
