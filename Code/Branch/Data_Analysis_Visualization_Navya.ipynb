```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Here are the relative paths for assorted resource folders. Feel free to save to a file within these paths. 
APIDataPath = '../Resources/API_Data/'
AnalyzedDataPath = '../Resources/Data_Analysis/'
FiguresPath = '../Resources/Figures/'
#e.g. TVdata = f'{APIDataPath}TV_dat84_23.csv'
```

```python
#Relative path
Electricity_df_path = '../Resources/API_Data/Cleaned_Electric_2013.csv'
```

```python
# Import the data from CSV
Electricity_df = pd.read_csv(Electricity_df_path)
Electricity_df.head()
```

```python
#converting Date coloumn into datetime formate
Electricity_df['Date'] = pd.to_datetime(Electricity_df['Date'])
#setting date coloumn as index
Electricity_df =Electricity_df.set_index('Date')
#create a figure and axis object
Fig, ax =plt.subplots(figsize=(10,5))
# Line Plot with Date and value
ax.plot(Electricity_df['Value'],linewidth=1.5, color= "blue")
#set the labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Value')
ax.set_title("Electricity Value change over time")
#display the graph
plt.show()
```

```python
# Create a histogram of the data
sns.histplot(data=Electricity_df, x='3-Month % Change', kde=True).set(title= "Electricity")
```

```python
# Calculate and print the standard deviation of the data
std_dev = np.std(Electricity_df['3-Month % Change'])
print(f"The standard deviation of the data is: {std_dev:.2f}")
```
